```python
# -*- coding: utf8

import sys
sys.path.append('../code/')
```

```python
from amutils import build_graph
from amutils import build_reverse_index
from amutils import load_am_json_data

from deep_features import features_and_pairwise

from scipy.stats import kendalltau
from scipy.stats import pearsonr
from scipy.stats import spearmanr

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
json_data = load_am_json_data()
feats_df, names_df, S = features_and_pairwise()
```

```python
store = pd.HDFStore('../cache/dt_matrices_new.h5')
DT = store['song_dt']
store.close()
```

```python
valid = S.index.intersection(DT.index)
S = S.loc[valid]
S = S[valid.values]
DT = DT.loc[valid]
DT = DT[valid.values]

assert(DT.shape == S.shape)
```

```python
_6M = pd.Timedelta(6, 'M').to_timedelta64().astype('timedelta64[s]').astype('f')
A = S[DT > _6M]
```

```python
song2artist = {}
for spotify_id, _, artist in names_df.values:
    song2artist[spotify_id] = artist
```

```python
artist_names = []
for song in A.columns:
    artist = song2artist[song]
    artist_names.append(artist)
```

```python
future_df = pd.DataFrame(A.mean(axis=0, skipna=True),
                         index=A.index.values,
                         columns=['sim'])
past_df = pd.DataFrame(A.mean(axis=1, skipna=True),
                       index=A.index.values,
                       columns=['sim'])

disrupt_song = (future_df - past_df) / (future_df + past_df)
disrupt_song['artist'] = artist_names
disrupt_artist = disrupt_song.groupby('artist').mean()['sim'].dropna()
```

```python
disrupt_artist.sort_values()
```

```python
valid = set(disrupt_artist.index)
keys_to_consider = set()
reverse_index = {}
for artist in json_data.keys():
    if json_data[artist]['name'] in valid:
        keys_to_consider.add(artist)
        reverse_index[json_data[artist]['name']] = artist
```

```python
G = build_graph(json_data, keys_to_consider, True)
from disrupt import compute_disruption
D = compute_disruption(G, 3, 3).dropna()
```

```python
D.sort_values(by='disruption')
```

```python
top_nodes = sorted(D['disruption'].iteritems(), key=lambda x: x[1], reverse=True)
top_similar = []
for name in disrupt_artist.sort_values()[::-1].index:
    if name in reverse_index:
        top_similar.append((reverse_index[name], disrupt_artist.loc[name]))
top_nodes = dict(top_nodes)
top_similar = dict(top_similar)
```

```python
x = []
y = []
valid = set(top_nodes.keys()).intersection(top_similar.keys())
for key in valid:
    x.append(top_nodes[key])
    y.append(top_similar[key])
x = np.array(x)
y = np.array(y)
```

```python
pearsonr(x, y)
```

```python
kendalltau(x, y)
```

```python
spearmanr(x, y)
```

```python
# Sanity plots from here on
plt.scatter(x, y)
```
